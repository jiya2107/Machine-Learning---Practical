```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Create Synthetic Data (House Price Example)
np.random.seed(42)
X = 2 * np.random.rand(100, 1) # Feature: Size (normalized)
y = 4 + 3 * X + np.random.randn(100, 1) # Target: Price (Linear relation + noise)

# 2. Gradient Descent Implementation
def gradient_descent(X, y, learning_rate=0.01, n_iterations=1000):
    # Add x0 = 1 to each instance for the intercept term (Bias)
    X_b = np.c_[np.ones((len(X), 1)), X]
    m = len(X) # Number of training examples

    # Initialize parameters (theta: [theta_0, theta_1] where theta_0 is intercept)
    theta = np.random.randn(2, 1)

    # Gradient Descent loop
    for iteration in range(n_iterations):
        # Calculate the hypothesis (h_theta(x))
        y_predicted = X_b.dot(theta)

        # Calculate the error (difference between predicted and actual)
        error = y_predicted - y

        # Calculate the gradient (partial derivative of MSE cost function)
        gradients = 2/m * X_b.T.dot(error)

        # Update theta
        theta = theta - learning_rate * gradients

    return theta

# Run Gradient Descent
final_theta = gradient_descent(X, y, learning_rate=0.1, n_iterations=1000)

print("--- Gradient Descent Results (Synthetic Data) ---")
print(f"Optimal Parameters (Intercept, Coefficient):\n{final_theta.flatten()}")

# Expected analytical solution for comparison: [4, 3]
```
